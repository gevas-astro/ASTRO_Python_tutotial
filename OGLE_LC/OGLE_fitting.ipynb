```python
#Lets plot some photometry data from OGLE survey for SMC X-2
# Download phot.dat file from following page
# http://ogle.astrouw.edu.pl/ogle4/xrom/sxp2.37.html

#load OGLE IV, only keep first three columns
og4 = np.loadtxt('phot2.dat',usecols=range(0,3))
#turn into panda dataframe with labelled columns based on known time units (Julian Date)
og4 = pd.DataFrame(og4,columns=['JD','I mag','I mag err'])

#add columns for MJD and MJD-50000, which will be the time unit used in all plots
og4['MJD'] = og4['JD']-2400000.5
og4['MJD-50000'] = og4['MJD']-50000
og4['I mag'] = og4['I mag']+.041 #calibrating
#deleting duplicates (points with same time)
og4 = og4.drop_duplicates(subset=['MJD'], keep='last')
```

```python
#Lets feet the green data with some function.

#thy a linear of parabolic function, but perhaps you can add more temrs, 
# like c*x**2 and d*x with c and d free parapeters... quad(a,x,b,c,d)

def parabola(x, a, b, c):
    x = np.array(x)
    return a*x**2 + b*x + c

def lin(x, a, b):
    x = np.array(x)
    return a*x + b
```

```python
fit_params, pcov = scipy.optimize.curve_fit(lin, og4m2['MJD'], og4m2['I mag'])
fit_params

y_fit2 = lin(og4m2['MJD'], *fit_params)
```

```python
# lets add errros/uncertainties in our optimizer.. og4m2['I mag err']
fit_params, pcov = scipy.optimize.curve_fit(parabola, og4m2['MJD'], og4m2['I mag'],sigma=og4m2['I mag err'])
fit_params

y_fit = parabola(og4m2['MJD'], *fit_params)
```
